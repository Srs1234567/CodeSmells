```python
import csv
```

```python
def createProject(name,month):
    projectName=''
    ending='</rdf:RDF>'
    beginning=['<?xml version="1.0"?>\n',
    '<rdf:RDF xmlns="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#"\n'.format(month,name),
    '     xml:base="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}"\n'.format(month,name),
    '     xmlns:owl="http://www.w3.org/2002/07/owl#"\n',
    '     xmlns:rdf="http://www.w3.org/1999/02/22-rdf-syntax-ns#"\n',
    '     xmlns:xml="http://www.w3.org/XML/1998/namespace"\n',
    '     xmlns:xsd="http://www.w3.org/2001/XMLSchema#"\n',
    '     xmlns:rdfs="http://www.w3.org/2000/01/rdf-schema#"\n',
    '     xmlns:Fitness_Package_Ricardo="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#">\n'.format(month,name),
    '    <owl:Ontology rdf:about="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}"/>\n'.format(month,name)]
    for row in beginning:
        projectName+=row
    return projectName, ending
```

```python
def getCsvData(link):
    dataList=[]
    with open(link) as csvFile:
        spamreader = csv.reader(csvFile)
        for row in spamreader:
            dataList.append(row)
    return dataList
```

```python
def createClasses(data,classes,month,name):
    for item in classes:
        if not isinstance(item, int):
            raise TypeError("List should contain only integers")
    classesDeclaration=[]
    classesName=[]
    for i in classes:
        className=data[i]
        classesName.append(className)
        rdf='<owl:Class rdf:about="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#{}"/>\n'.format(month,name,className)
        classesDeclaration.append(rdf)
    return classesName, classesDeclaration
```

```python
def createObjectProperties(month,name,objProperties):
    propertiesList=[]
    for class1, classes2 in objProperties.items():
        for class2, properties in classes2.items():
            newProperty='<owl:ObjectProperty rdf:about="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#{}"/>\n'.format(month, name, properties)
            propertiesList.append(newProperty)
    return propertiesList
```

```python
def createDataProperties(month,name,dataProperties,columnNames):
    propertiesList=[]
    
    for class1, properties in dataProperties.items():
        for dataX, rangeX in properties.items():
            value=columnNames[dataX]
            value = value.replace(" ", "_")
            firstL='<owl:DatatypeProperty rdf:about="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#{}">\n'.format(month,name,value)
            secondL='    <rdfs:range rdf:resource="http://www.w3.org/2001/XMLSchema#{}"/>\n'.format(rangeX)
            thirdL='</owl:DatatypeProperty>\n'
            propertiesList.append(firstL+secondL+thirdL)
    return propertiesList
```

```python
def createInstances(position, classesNames, toInstance, month, name, dataProperties, objectProperties):
    classNum=0
    instances=[]
    for j in position:
        #get current class
        className=classesNames[classNum]
        for vals in toInstance:
            #get object property
            objProps=[]
            
            for class2, objProp in objectProperties[j].items():
                val=vals[class2]
                val=val.replace(" ", "_")
                objPropDec='       <{} rdf:resource="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#{}"/>\n'.format(objProp, month,name,val)
                objProps.append(objPropDec)

            
            #get data property
            dataProps=[]
            for class2, dataProp in dataProperties[j].items():
                prop=classesNames[class2]
                prop=prop.replace(" ", "_")
                rang=dataProp
                val=vals[class2]
                val=val.replace(" ", "_")
                dataPropDec='       <{} rdf:datatype="http://www.w3.org/2001/XMLSchema#{}">{}</{}>\n'.format(prop,rang,val,prop)
                dataProps.append(dataPropDec)

                
            #make instance
            instanceName=vals[j]
            instanceName = instanceName.replace(" ", "_")
            instanceList=['   <owl:NamedIndividual rdf:about="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#{}">\n'.format(month,name,instanceName),
                         '       <rdf:type rdf:resource="http://www.semanticweb.org/usuario/ontologies/2023/{}/{}#{}"/>\n'.format(month,name,className),
                         ]
            instance=''
            for i in instanceList:
                instance+=i
            for i in objProps:
                instance+=i
            for i in dataProps:
                instance+=i
            instance+='   </owl:NamedIndividual>\n'
            instances.append(instance)
        classNum+=1
    return instances
```

```python
def createDocFile(fileName, declaration):
    with open(fileName, 'w') as f:
        for row in declaration:
            if row is not str:
                for row2 in row:
                    f.write(row2)
                    
                
            else:
                f.write(row)
                f.write('\n')
            f.write('\n')
```

```python
def get_data_props(val):
    link='datafiles/final.csv'
    data=getCsvData(link)
    cn=data[0]
    dataProperties={'Person':{cn[4]:'string',cn[5]:'integer'},'Company':{cn[6]:'string',cn[7]:'string'},'City':{cn[3]:'float'}}
    return dataProperties[val]

def all_data_props():
    link='datafiles/final.csv'
    data=getCsvData(link)
    cn=data[0]
    dataProperties={cn[4]:'string',cn[5]:'integer',cn[6]:'string',cn[7]:'string',cn[3]:'float'}
    return dataProperties
```

```python
def main():
    projectName='Proyecto'
    fileName='try1.owl'
    link='datafiles/final.csv'
    month=5
    classesPosition=[0,1,2]
    objProperties={0:{1:'worksAt',2:'livesIn'}, 1:{0:'hasEmployee',2:'situatedAt'},2:{0:'hasCitizen',1:'hasCompany'}}
    dataProperties={0:{4:'string',5:'integer'},1:{6:'string',7:'string'},2:{3:'float'}}
    
    #GET CSV DATA, SPLIT COLUMN NAMES AND INSTANCES 
    data=getCsvData(link)
    columnNames=data[0]
    instancesList=data[1:]
    
    #Create project
    proyectDeclaration, endDeclaration=createProject(projectName, month)
    print('Project Declaration was succesful')
    #Create classes
    classesName, classesDeclaration = createClasses(columnNames,classesPosition,month,projectName)
    print('Class Declaration was succesful')
    
    
    #Create object properties
    objProps=createObjectProperties(month, projectName, objProperties)
    
    #Create data properties
    dataProps=createDataProperties(month, projectName, dataProperties,columnNames)
    
    #Create instances
    instancesDeclaration=createInstances(classesPosition,columnNames, instancesList,month, projectName, dataProperties,objProperties)
    print('Instances Declaration was succesful')
    
    
    #Create docFile
    createDocFile(fileName, [proyectDeclaration,classesDeclaration,objProps,dataProps, instancesDeclaration,endDeclaration])
    print('File Creation was succesful')
    
    print('File Successfully Created')
```

```python
main()
```
